```python
def concat_path(path, series):
    series = path + series
    return series

train_df["filepath"] = concat_path("standford_cars/train/",train_df["filename"])
train_df.head()
```

```python
# check if the data frame files are equal to the data in the training sets
if len(os.listdir("standford_cars/train")) == len(train_df["filepath"]):
    print("number matches")
    
else:
    print("number does not match")
```

```python
train_df.label.describe()
```

```python
# The values are string and we cannot interpret using numpy 
# converting into the number as required 
# we have two ideas either use label encoding or One Hot Encoding
# Label Encoder has dis of assuming ranking 
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train_df.label = le.fit_transform(train_df.label)
```

```python
train_df
```

```python
# NEXT STEP is to check image size and shapes
```

```python
import cv2
image_size = 100
data = []
# lets read the image into grayscale and transform into the size 

for i in range(0,len(train_df)):
    img_array = cv2.imread(train_df.filepath[i], cv2.IMREAD_GRAYSCALE)
    new_img = cv2.resize(img_array, (image_size, image_size))
    data.append([new_img, train_df['label'][i]])
```

```python
np.random.shuffle(data)
```

```python
x = []
y = []
for image in data:
    x.append(image[0])
    y.append(image[1])

# converting x & y to numpy array as they are list
x = np.array(x)
y = np.array(y)
```

```python
np.unique(y, return_counts=True)
```

```python
# Now we have the data
# lets divide into train and test set 

x.shape
```

```python
x = x.reshape(-1, 100, 100, 1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
# split the data
X_train, X_val, y_train, y_val = train_test_split(x,y,test_size=0.3, random_state = 42)
```

```python
! pip install tensorflow 
```

```python
import tensorflow as tf

cnn = tf.keras.models.Sequential([
tf.keras.layers.Conv2D(filters=32, kernel_size=(3, 3), activation='relu', input_shape=(100, 100, 1)),
tf.keras.layers.MaxPooling2D((2, 2)),

tf.keras.layers.Conv2D(filters=64, kernel_size=(3, 3), activation='relu'),
tf.keras.layers.MaxPooling2D((2, 2)),

# tf.keras.layers.Flatten(input_shape=(100, 100, 1)),
tf.keras.layers.Flatten(),
tf.keras.layers.Dense(64, activation='relu'),
tf.keras.layers.Dense(131, activation='sigmoid')
])
```

```python
cnn.compile(optimizer='adam',
loss='sparse_categorical_crossentropy',
metrics=['accuracy'])
```

```python
cnn.fit(X_train, y_train, epochs=5, batch_size=2)
```

```python
cnn.evaluate(X_val, y_val)
```

```python
# Loading the order of the image's name that has been provided
test_image_order = pd.read_csv("standford_cars/Testing_set.csv")
test_image_order.head()
```

```python
test_image_order["filepath"] = concat_path("standford_cars/test/",test_image_order["filename"])
test_image_order.head()
```

```python
import cv2
image_size = 100
data = []
# lets read the image into grayscale and transform into the size 

for i in range(0,len(test_image_order)):
    img_array = cv2.imread(test_image_order.filepath[i], cv2.IMREAD_GRAYSCALE)
    new_img = cv2.resize(img_array, (image_size, image_size))
    data.append([new_img])
    
```

```python
x = []
y = []
for image in data:
    x.append(image[0])

# converting x & y to numpy array as they are list
x = np.array(x)
```

```python
x_pixel = x.reshape(-1, 100, 100, 1)
```

```python
pred  = cnn.predict(x_pixel)
```

```python
prediction = []
for value in pred:
    prediction.append(np.argmax(value))
```

```python
predictions = le.inverse_transform(prediction)
```

```python
res = pd.DataFrame({'filename': test_image_order['filename'], 'label': predictions}) 
res.to_csv("submission.csv", index = False)
```
